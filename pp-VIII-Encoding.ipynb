Orden de las variables categ칩ricas:
- Tienen orden: 'market_size','market' y 'a침o'. Utilizamos map
- No tienen orden: 'product_type'. Utilizamos get_dummies

```python
mapa_market = {'Central':1,'West': 2, 'East': 3, 'South': 4} # encoding sobre 'market'
```

```python
sales['market_encoding'] =sales['market'].map(mapa_market)
```

```python
sales.head(2)
```

```python
mapa_market_size = {'Major Market':1,'Small Market': 0} # encoding sobre 'market_size'
```

```python
sales['market_size_encoding'] =sales['market_size'].map(mapa_market_size)
```

```python
sales.head(2)
```

```python
mapa_a침o = {2010:0,2011:1} # encoding sobre a침o
```

```python
sales['a침o_encoding'] =sales['a침o'].map(mapa_a침o)
```

```python
sales.head(2)
```

```python
product_type_dummies = pd.get_dummies(sales['product_type'],dtype = int) # encoding sobre 'product_type
```

```python
df_dummies = pd.concat([sales,product_type_dummies], axis = 1)
```

```python
df_dummies.head(2)
```

```python
df_dummies.drop(columns = ['product','state'],axis = 1, inplace =True)
```
